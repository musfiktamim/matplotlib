```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
classes = ["python","R","Machine Learning","Artificial Intelligence","Data Science"]
class1_students = [45,15,35,25,30]
plt.pie(class1_students,labels=classes)
plt.show()
```

```python
plt.figure(figsize=(10,8))
classes = ["python","R","Machine Learning","Artificial Intelligence","Data Science"]
class1_students = [45,15,35,25,30]
plt.pie(class1_students,labels=classes,labeldistance=1.1,autopct="%0.2f%%")
plt.legend(classes,loc=4)
plt.show()
```
