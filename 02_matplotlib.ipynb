```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
ml_Student_age = np.random.randint(18,45,(100))
py_Student_age = np.random.randint(18,45,(100))
```

```python
print(ml_Student_age,py_Student_age)
```

```python
plt.hist(ml_Student_age)
plt.title("ml student age hist")
plt.xlabel("students age category")
plt.ylabel("no students age")
plt.show()
```

```python
plt.hist(ml_Student_age,bins=[15,20,25,30,35,40,45])
plt.title("ml student age hist")
plt.xlabel("students age category")
plt.ylabel("no students age")
plt.show()
```

```python
plt.hist(ml_Student_age,bins=[15,20,25,30,35,40,45],rwidth=0.9)
plt.title("ml student age hist")
plt.xlabel("students age category")
plt.ylabel("no students age")
plt.show()
```

```python
plt.hist(ml_Student_age,bins=[15,20,25,30,35,40,45],rwidth=0.9,histtype="bar")
plt.title("ml student age hist")
plt.xlabel("students age category")
plt.ylabel("no students age")
plt.show()
```

```python
plt.hist(ml_Student_age,bins=[15,20,25,30,35,40,45],rwidth=0.9,histtype="bar",orientation="horizontal")
plt.title("ml student age hist")
plt.xlabel("students age category")
plt.ylabel("no students age")
plt.show()
```

```python
plt.hist(ml_Student_age,bins=[15,20,25,30,35,40,45],rwidth=0.9,histtype="bar",orientation="horizontal",color="g")
plt.title("ml student age hist")
plt.xlabel("students age category")
plt.ylabel("no students age")
plt.legend(["Ml students"])
plt.show()
```

```python
plt.hist([ml_Student_age,py_Student_age],label=["ml students","py students"],rwidth=0.9,bins=[15,20,25,30,35,40,45])
plt.legend()
plt.show()
```

```python
from matplotlib import style
style.use("ggplot")
plt.grid()
plt.hist([ml_Student_age,py_Student_age],label=["ml students","py students"],rwidth=0.9,bins=[15,20,25,30,35,40,45])
plt.legend()
plt.show()
```
