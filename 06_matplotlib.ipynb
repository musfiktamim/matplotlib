```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
classes = ["Python","R","Machine Learning","Artificial Intelligence","Data Science"]
class1_students = [45,15,35,25,30]
display(classes,class1_students)
```

```python
plt.pie([1])
plt.show()
```

```python
plt.pie(class1_students,labels=classes)
plt.show()
```

```python
explode = [0.03,0,0.2,0,0]
color = ["c","b","r","y","g"]
plt.pie(class1_students,labels=classes,explode=explode,colors=color,autopct="%0.3f",shadow=True,radius=1.4,startangle=)
plt.show()
```
